```python
from __future__ import division
from matplotlib import pyplot as plt
from numpy import *
%matplotlib inline
```

```python
freq=100.0          #signal frequency
```

```python
#original signal
fs=10000           # considering sampling frequency much higher
plt.figure(figsize=(7,3))
ot=arange(-1/100,2/100+1/fs,1/(2*fs))
func=sin(2*pi*freq*ot)
plt.plot(ot,func) 
plt.title("Original Signal")
plt.xlabel('Time',fontsize=8)
plt.ylabel('Amplitude',fontsize=8)
plt.grid(True)
```

```python
for i in range(2,20):
    fs= i*50       #sampling frequency incrementing in multiple of 50   
    t=arange(-1/100,2/100+1/fs,1/fs)
    x = sin(2*pi*freq*t)
    #print 'value of i//2 is'+str(i//2-1)+'value of i%2 is'+str(i%2)
    plt.figure(figsize=(7,3))
    
#     plt.subplot(10,1,i-1)
    plt.plot(t,x,'o-')
    plt.subplots_adjust(hspace=0)
    if(fs<2*freq):
        plt.title('under sampled, fs is '+str(fs)+', freq ='+str(freq),fontsize=8)
    else:       
        plt.title('Fs is '+str(fs)+', freq ='+str(freq),fontsize=8)
    plt.xlabel('Time',fontsize=8)
    plt.ylabel('Amplitude',fontsize=8)
    plt.grid(True)
```
